```python
# Join ratings and books dataframes
df=pd.merge(df_ratings,df_books,on="ISBN")

# Shape of the data
df.shape
```

```python
# Take 1 % data as sample  
df_sample = df.sample(frac=.01, random_state=1) 

# Shape of the sample data
df_sample.shape
```

```python
# Create Item-user matrix using pivot_table()
df_pivot = df_sample.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating').fillna(0)

# Show top-5 records
df_pivot.head()
```

```python
# Import NearestNeighbors
from sklearn.neighbors import NearestNeighbors

# Build NearestNeighbors Object
model_nn = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=7, n_jobs=-1)

# Fit the NearestNeighbor
model_nn.fit(df_pivot)
```

```python
# Get top 10 nearest neighbors 
indices=model_nn.kneighbors(df_pivot.loc[['Souls Belated (Penguin Classics 60s)']], 10, return_distance=False)

# Print the recommended books
print("Recommended Books:")
print("==================")
for index, value in enumerate(df_pivot.index[indices][0]):
    print((index+1),". ",value)
```
